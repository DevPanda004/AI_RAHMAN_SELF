```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer, pipeline
# Load pre-trained model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
# Setup pipeline
gpt2_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# Generate text
prompt = "Once upon a time"
generated_texts = gpt2_pipeline(prompt, max_length=50, num_return_sequences=1)
print(generated_texts)
```

```python
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import requests
from io import BytesIO

# Load CLIP model and processor
model_name = "openai/clip-vit-base-patch32"
model = CLIPModel.from_pretrained(model_name)
processor = CLIPProcessor.from_pretrained(model_name)
```

```python
# Load an example image from a direct link
url = "https://images.unsplash.com/photo-1518791841217-8f162f1e1131"  # Direct image link

response = requests.get(url)
if response.status_code == 200:
    image = Image.open(BytesIO(response.content))
else:
    raise Exception("Failed to fetch image")

# Prepare inputs
inputs = processor(images=image, return_tensors="pt")

# Define the labels you want to classify the image with
labels = ["a photo of a cat", "a photo of a dog", "a photo of a bird"]

# Tokenize the labels
text_inputs = processor(text=labels, return_tensors="pt", padding=True)

# Forward pass
outputs = model(pixel_values=inputs.pixel_values, input_ids=text_inputs.input_ids, attention_mask=text_inputs.attention_mask)
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)  # Convert logits to probabilities

print("Probabilities:", probs)
```
